**MULTIVARIATE LINEAR REGRESSION PREDICTING HOUSE PRICE USING MRT STATION, CONVENIENCE STORE AND HOUSE AGE AS INDEPENDENT VARIABLES**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Real Estate - Real Estate (1).csv')
```

```python
df= pd.read_csv('/content/Real Estate - Real Estate (1).csv')

print(df.to_string())
```

```python
df.info()
```

```python
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python

x = df[['X3 distance to the nearest MRT station','X4 number of convenience stores','X2 house age']]

y= df[['Y house price of unit area']]

# Taking only the selected two attributes from the dataset
x.columns = [ 'MRT STATION','CONVENIENCE STORE','HOUSE AGE']

y.columns=['HOUSE PRICE']

```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(x,y)
y_pred=l.predict(x)
print(y_pred)
```

```python
from sklearn.metrics import r2_score  ##The r2_score function is part of the sklearn.metrics module in scikit-learn, a popular machine learning library in Python. It is used to calculate the coefficient of determination, also known as R-squared, which is a metric to evaluate the performance of regression models.
r2_score(y, y_pred)
```

**MULTIVARIATE LINEAR REGRESSION PREDICTING HOUSE PRICE USING Latitude, Longitude AND Transaction Date AS INDEPENDENT VARIABLES**

```python

x = df[['X5 latitude','X6 longitude','X1 transaction date']]

y= df[['Y house price of unit area']]

# Taking only the selected two attributes from the dataset
x.columns = [ 'LAT','LONG','TRANSC DATE']

y.columns=['HOUSE PRICE']

```

```python
l=LinearRegression()
l.fit(x,y)
y_pred=l.predict(x)
print(y_pred)
```

```python
from sklearn.metrics import r2_score  ##The r2_score function is part of the sklearn.metrics module in scikit-learn, a popular machine learning library in Python. It is used to calculate the coefficient of determination, also known as R-squared, which is a metric to evaluate the performance of regression models.
r2_score(y, y_pred)
```
